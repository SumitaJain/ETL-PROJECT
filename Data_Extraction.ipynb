```python
# Dependencies
import pandas as pd
import os
import csv
```

```python
# load CSV's
total_toll= os.path.join("total-toll.csv")
total_toll_non_interstate =os.path.join("total-toll-nis.csv")
road_congestion_index=os.path.join("Road Congestion index.csv")
state_tollroads_bridges=os.path.join("state-toll-roads-bridges.csv")
state_tollroads_bridges_2010=os.path.join("state-toll-roads-bridges-2010.csv")
```

```python
# Read with pandas
state_tolls_2012_df=pd.read_csv(state_tollroads_bridges)
state_tolls_2010_df=pd.read_csv(state_tollroads_bridges_2010)
total_toll_df=pd.read_csv(total_toll)
road_congestion_index_df=pd.read_csv(road_congestion_index)
total_toll_noninterstate_df=pd.read_csv(total_toll_non_interstate)
```

```python
state_tolls_2012_df=state_tolls_2012_df[["State", "Toll road mileage", "Number of toll bridges", "Number of toll tunnels"]]
state_tolls_2012_df["Year"]=2012
state_tolls_2012_df.head()
```

```python
state_tolls_2010_df=state_tolls_2010_df[["State", "Toll road mileage", "Number of toll bridges", "Number of toll tunnels"]]
state_tolls_2010_df["Year"]=2010
state_tolls_2010_df.head()
```

```python
#Merging state and associated toll data for 2010 and 2012 into one dataframe

state_tolls_combined_df=state_tolls_2010_df.append(state_tolls_2012_df)
state_tolls_combined_df.head()
```

```python
road_congestion_index_df=road_congestion_index_df[["Urban area", "Population group", "2011"]]
road_congestion_index_df.head()
```

```python
#Splitting city and state from road congestion index

road_congestion_index_df["City"]=""
road_congestion_index_df["State"]=""
for i in range(len(road_congestion_index_df)):
    state=road_congestion_index_df["Urban area"][i][-4:]
    city=road_congestion_index_df["Urban area"][i].replace(state, '')
    state=state[-2:]
    road_congestion_index_df["City"][i]=city
    road_congestion_index_df["State"][i]=state
road_congestion_index_df.head()
```

```python
#Removing combined city state column
road_congestion_index_df=road_congestion_index_df[["State", "City", "Population group", "2011"]]
road_congestion_index_df.head()
```
